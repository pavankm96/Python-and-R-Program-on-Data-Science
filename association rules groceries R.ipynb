```r
library(arules)
```

```r
groceries=read.csv("Datasets/groceries.csv")
```

```r
groceries_rules=apriori(groceries,parameter = list(support=0.002,confidence=0.8,minlen=2))
```

```r
library(arulesViz)
```

```r
plot(groceries_rules,method = "scatterplot",jitter=0)
```

```r
plot(groceries_rules,method = "grouped")
```

```r
plot(groceries_rules,method = "graph")
```

```r
groceries_conf=sort(groceries_rules,by="confidence")
inspect(head(groceries_conf))
```

```r
groceries_lift=sort(groceries_rules,by="lift")
inspect(head(groceries_lift))
```
