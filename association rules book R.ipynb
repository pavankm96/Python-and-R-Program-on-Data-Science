```r
library(arules)
```

```r
book=read.csv("Datasets/book.csv")
```

```r
book_rules=apriori(as.matrix(book),parameter = list(support=0.002,confidence=0.5,minlen=2))
```

```r
library(arulesViz)
```

```r
plot(book_rules,method = "scatterplot",jitter=0)
```

```r
plot(book_rules,method = "grouped")
```

```r
plot(book_rules,method = "graph")
```

```r
book_conf=sort(book_rules,by="confidence")
inspect(head(book_conf))
```

```r
book_lift=sort(book_rules,by="lift")
inspect(head(book_lift))
```
